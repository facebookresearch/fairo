# Video Viewer for rgb based datasets

```python
import subprocess
import os

root = '/scratch/apratik/dfeb/locobot/robot/finals'
# root = '/checkpoint/apratik/finals'
# root = '/checkpoint/apratik/ActiveVision/active_vision/replica_random_exploration_data'
scene = 'office_2'
heu = 'straightline'
vid_name = scene + '_' + heu
# rgb_path = '/scratch/apratik/dfeb/locobot/slam_logs/apartment_0/rgb' 
rgb_path = os.path.join(root, heu, scene, 'rgb')
out_f = os.path.join(os.getcwd(), f'videos/finals/{vid_name}.mp4')
print(out_f)
```

```python
if os.path.exists(out_f):
    os.remove(out_f)
```

```python
# command = f"ffmpeg -framerate 6 -f image2 -i {os.path.join(root, scene, 'rgb')}/%05d.jpg {out_f}"
command = f"ffmpeg -framerate 50 -i {rgb_path}/%05d.jpg -c:v libx264 -vf 'fps=25,format=yuv420p' {out_f}"
print(command)
```

```python
subprocess.call(command, shell=True)
```
